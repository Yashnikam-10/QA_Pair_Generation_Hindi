````python
import datasets
import json
import torch
import pandas as pd
````

````python
df = datasets.load_dataset("ai4bharat/IndicQA", "indicqa.hi")
````

````python
df = df['test']
````

````python
df_hindi = pd.DataFrame(df)
````

````python
df_hindi.head(10)
````

````python
li = []
for i in range(len(df_hindi.answers)):
    if df_hindi.answers[i]['text'][0] == '':
        li.append(i)
````

````python
print(len(li))
````

````python
df_hindi = df_hindi.drop(li)
````

````python
df_hindi.head(10)
````

````python
print(len(df_hindi.context.unique()))
````

````python
df_hindi = df_hindi.reset_index()
````

````python
essay_dict = {}
for essay in df_hindi.context.unique():
    essay_dict[essay] = []

# print(essay_dict)
````

````python
print(df_hindi.iloc[4]['context'])
````

````python
for i in range(df_hindi.shape[0]):
    essay_dict[df_hindi.iloc[i]['context']].append(i)
````

````python
for essay in df_hindi.context.unique():
    li = essay_dict[essay]
    qa_prompt = """"""
    for i, idx in list(zip(li, [i+1 for i in range(len(li))])):
        ques = df_hindi.iloc[i]['question']
        ans = df_hindi.iloc[i]['answers']['text'][0]
        
        app = f"Question {idx}: " + str(ques) + " " + f"Answer {idx}: " + str(ans)+ " "
        qa_prompt += str(app)
    essay_dict[essay] = qa_prompt
````

````python
context_list = list(essay_dict.keys())
qa_list = list(essay_dict.values())
````

````python
df_qa = pd.DataFrame({'context': context_list, 'qa': qa_list})
df_qa.head()
````

````python
def buildprompt(data):
    prompt = {}
    prompt['text'] = "Given the context delimited by triple backticks ```" + data['context'] + "```, generate some question answer pairs. Questions and Answers: " + data['qa'] + "" 
    return prompt

df_qa['prompt'] = df_qa.apply(buildprompt, axis=1)

result = df_qa['prompt'].to_list()
````
